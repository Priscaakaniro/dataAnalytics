```python
import pandas as pd

retailData = pd.read_csv('online_retail.csv')

retailData.head(10)
```

```python
retailData.info()
```

```python
retailData.isna().sum()
```

Cell above shows the columns and how many missing values per column. Will begin EDA now

```python
retailData.replace(['wet?','Discount', 'counted', '?lost', 'mixed up', '? sold as sets?','?sold as sets?', 'Given away', 'label mix up', 'samples/damages', 'thrown away', 'showroom' 'missing','????damages????', 'wet boxes', 'wet', 'lost'], pd.NA, inplace=True)
```

```python
mode = retailData['Description'].mode()[0]
retailData['Description'] = retailData['Description'].fillna(mode)
```

```python
retailData['Description'].unique()
```

```python
median = retailData['CustomerID'].median()
retailData['CustomerID'] = retailData['CustomerID'].fillna(median)
```

```python
retailData['CustomerID'].unique()
```

```python
retailData.isna().sum()
```

Changing data types

```python
retailData['InvoiceDate'] = pd.to_datetime(retailData['InvoiceDate'])

retailData['InvoiceDate'] = retailData['InvoiceDate'].dt.date
```

```python
retailData['CustomerID'] = retailData['CustomerID'].astype(int)
```

```python
retailData.info()
```

What country do people buy from the most?

```python
retailData['Country'].sort_values(ascending=False)
#Italy
```

Total price for the products

```python
retailData['TotalPrice'] = retailData['Quantity'] * retailData['UnitPrice']
retailData.head()
```

extracting month and year from invoice date

```python
retailData['Month'] = pd.to_datetime(retailData['InvoiceDate']).dt.month
retailData['Year'] = pd.to_datetime(retailData['InvoiceDate']).dt.year
retailData.head()
```

```python
retailData= retailData.drop(labels=['InvoiceYear','InvoiceMonth'], axis=1)
retailData.head()
```

What year has the most orders

```python
retailData['Year'].sort_values(ascending=False)
#2011
```

Saving as CSV

```python
retailData.to_csv("retailData_cleaned.csv", index=False)
```
